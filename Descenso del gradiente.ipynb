```python
import numpy as np
import scipy as sc

import matplotlib.pyplot as plt
```

###### Formula de la ecuaci√≥n del plano $$F{(x,y)} = sin({-1 \over 2}x^{2} - {1 \over 4}y^{2} + 3)cos(2x + 1 - e^{y})$$


```python
#theta = referencia al parametro de la funci√≥n anonima(lambda)
func = lambda th: np.sin(1/2 * th[0] ** 2 -1/4 * th[1] **2 + 3) * np.cos(2* th[0] + 1 - np.e ** th[1])

res = 100

_X = np.linspace(-5, 5, res) ##Llenar X
_Y = np.linspace(-5, 5, res) ##Llenar Y
_Z = np.zeros((res, res)) ##Crear Z

for ix, x in enumerate(_X): ##Llenar Z
    for iy, y in enumerate(_Y):
        _Z[iy, ix] = func([x, y])
        
plt.contourf(_X, _Y, _Z, 100)
plt.colorbar()


Theta = np.random.rand(2) * 4 - 2 ##Theta es un punto aleatorio en el gradiente

_T = np.copy(Theta)

h = 0.001                  ## diferencial en la derivada
lr = 0.01               ## Ratio de aprendizaje

plt.plot(Theta[0], Theta[1], "o", c="white")

grad = np.zeros(2)

for _ in range(100000):
    for it, th in enumerate(Theta): ##Derivada parcial de theta
        
        _T = np.copy(Theta)
        
        _T[it] = _T[it] + h
    
        deriv = (func(_T) - func(Theta)) / h
    
        grad[it] = deriv
        
    Theta = Theta - lr * grad
    
    ##print(func(Theta)) 

    if(_ % 100 == 0) :
        plt.plot(Theta[0], Theta[1], ".", c="red")
    
plt.plot(Theta[0], Theta[1], "o", c="black")

plt.show()
```
